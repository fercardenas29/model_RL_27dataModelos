# Cargamos nuestra Data a un Df

#### Librerias

```python
import os
import pandas as pd
```

#### Primer df 
1. Analizando la estructura que tiene nuestra Data, podemos identificar:

        Es un directorio que contiene 27 archivos txt
        En el nombre de cada archivo tenemos el punto (coordenada) donde se saco la muestra
        Dentro de cada txt tenemos 2 columanas, que equivalen a valores de energía

2. Vamos a contruir nuestro df con el siguiente flujo y estructura

        Extraer coordenadas del nombre del archivo (ejemplo: "DatReceptor_0_0_1.txt")
        Sumar la columna 1 y columna 2 fila por fila (columna0 + columna1)
        Estructura de las columndas del df =[x,y,z,valor]

```python


# Ruta de la carpeta donde están los archivos .txt
carpeta = "/home/fcardenas/Documentos/OctavoSemestre/modelos/27Datos/Data"

# Crear una lista para almacenar los datos procesados
data = []

# Leer todos los archivos .txt en la carpeta
for archivo in os.listdir(carpeta):
    if archivo.endswith(".txt"):
        # Extraer coordenadas del nombre del archivo (ejemplo: "DatReceptor_0_0_1.txt")
        nombre_base = os.path.splitext(archivo)[0]  # Eliminar la extensión .txt
        partes = nombre_base.split("_")
        x, y, z = map(int, partes[1:])  # Convertir las coordenadas a enteros

        # Leer el archivo como un DataFrame
        file_path = os.path.join(carpeta, archivo)
        datos = pd.read_csv(file_path, delim_whitespace=True, header=None)

        # Sumar fila por fila (columna 0 + columna 1)
        suma_filas = datos[0] + datos[1]

        # Agregar las coordenadas y los valores al listado
        for valor in suma_filas:
            data.append([x, y, z, valor])

# Crear un DataFrame con las columnas 'x', 'y', 'z' y 'valor'
df1 = pd.DataFrame(data, columns=["x", "y", "z", "valor"])
print(df1)
```

#### Segundo df 

1. Vamos a contruir nuestro df en base al analisis del primero pero con una diferente estructura

        Extraer coordenadas del nombre del archivo (ejemplo: "DatReceptor_0_0_1.txt")
        Sumar la columna 1 y columna 2 fila por fila (columna0 + columna1)
        Estructura de las columndas del df =[x,y,z,valor1.....valor200], dado que aqui vamos a trabajar con los 200 primeros registros de cada coordenada

```python
# Ruta de la carpeta donde están los archivos .txt
carpeta = "/home/fcardenas/Documentos/OctavoSemestre/modelos/27Datos/Data"

# Crear una lista para almacenar los datos procesados
data = []

# Leer todos los archivos .txt en la carpeta
for archivo in os.listdir(carpeta):
    if archivo.endswith(".txt"):
        # Extraer coordenadas del nombre del archivo (ejemplo: "DatReceptor_0_0_1.txt")
        nombre_base = os.path.splitext(archivo)[0]  # Eliminar la extensión .txt
        partes = nombre_base.split("_")
        x, y, z = map(int, partes[1:])  # Convertir las coordenadas a enteros

        # Leer el archivo como un DataFrame
        file_path = os.path.join(carpeta, archivo)
        datos = pd.read_csv(file_path, delim_whitespace=True, header=None)

        # Sumar fila por fila (columna 0 + columna 1)
        suma_filas = datos[0] + datos[1]

        # Tomar solo los primeros 200 valores (si existen)
        suma_filas_limited = suma_filas.head(200)

        # Convertir los valores en una lista y agregarla al listado con coordenadas
        data.append([x, y, z] + suma_filas_limited.tolist())

# Crear el DataFrame
columnas = ["x", "y", "z"] + [f"valor_{i+1}" for i in range(len(suma_filas_limited))]
df = pd.DataFrame(data, columns=columnas)

# Mostrar el DataFrame
print(df)
```

#### Guardar df en Disco como archivo .csv

```python
df.to_csv("dataframe2.csv", index=False)
```
