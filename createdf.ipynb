```python
# Ruta de la carpeta donde están los archivos .txt
carpeta = "/home/fcardenas/Documentos/OctavoSemestre/modelos/27Datos/Data"

# Crear una lista para almacenar los datos procesados
data = []

# Leer todos los archivos .txt en la carpeta
for archivo in os.listdir(carpeta):
    if archivo.endswith(".txt"):
        # Extraer coordenadas del nombre del archivo (ejemplo: "DatReceptor_0_0_1.txt")
        nombre_base = os.path.splitext(archivo)[0]  # Eliminar la extensión .txt
        partes = nombre_base.split("_")
        x, y, z = map(int, partes[1:])  # Convertir las coordenadas a enteros

        # Leer el archivo como un DataFrame
        file_path = os.path.join(carpeta, archivo)
        datos = pd.read_csv(file_path, delim_whitespace=True, header=None)

        # Sumar fila por fila (columna 0 + columna 1)
        suma_filas = datos[0] + datos[1]

        # Tomar solo los primeros 200 valores (si existen)
        suma_filas_limited = suma_filas.head(200)

        # Convertir los valores en una lista y agregarla al listado con coordenadas
        data.append([x, y, z] + suma_filas_limited.tolist())

# Crear el DataFrame
columnas = ["x", "y", "z"] + [f"valor_{i+1}" for i in range(len(suma_filas_limited))]
df = pd.DataFrame(data, columns=columnas)

# Mostrar el DataFrame
print(df)
```

# Pasos para realizar la interpolación

```python
from scipy import interpolate
import numpy as np
```

```python
# Las coordenadas (x, y, z) y los valores asociados
points = np.array([(i, j, k) for i in range(len(grid_x)) for j in range(len(grid_y)) for k in range(len(grid_z))])

# Los valores para cada punto (valor_1, valor_2, ..., valor_200) en una lista
values = values_matrix_200.reshape(-1, 200)
values_reshaped = values.reshape(len(grid_x), len(grid_y), len(grid_z), 200)
```

¿Por qué esta estructura es necesaria?

El interpolador espera una estructura de datos que esté organizada de manera coherente con las coordenadas de la malla xx, yy, y zz, junto con los valores correspondientes. Al tener un arreglo 4D, se asegura que cada valor de xx, yy, y zz tenga su propio conjunto de 200 valores asociados, lo que te permite interpolar entre esas 200 características para cada punto en el espacio 3D definido por xx, yy, y zz.

```python
print(values_reshaped)
```

```python
from scipy.interpolate import RegularGridInterpolator

# Suponiendo que values_reshaped tiene la forma (len(grid_x), len(grid_y), len(grid_z), 200)
interpolators = []
for l in range(200):  # Para cada uno de los 200 valores
    # Crear el interpolador para el valor específico
    interp = RegularGridInterpolator((grid_x, grid_y, grid_z), values_reshaped[:, :, :, l], method='linear', bounds_error=False, fill_value=None)
    interpolators.append(interp)
```

Las coordenadas grid_x, grid_y y grid_z forman una malla regular, podrías seleccionar puntos dentro de esta malla regular para realizar la interpolación. En lugar de hacerlo con puntos arbitrarios, seleccionaremos puntos dentro de las coordenadas definidas en grid_x, grid_y y grid_z.
Pasos para la interpolación en una malla regular:

    Definir puntos de interpolación dentro de la malla regular: Si quieres interpolar en puntos dentro de las coordenadas que ya tienes (sin exceder los límites de la malla), puedes elegir puntos dentro de grid_x, grid_y y grid_z. Por ejemplo, puedes elegir puntos en una submalla o realizar un muestreo de manera regular.

    Utilizar la interpolación: Después, puedes interpolar esos puntos dentro de los límites de la malla regular utilizando el RegularGridInterpolator como lo mencionamos anteriormente.

Voy a mostrarte cómo seleccionar puntos dentro de la malla regular y hacer la interpolación:
Código para la interpolación en puntos dentro de la malla regular:

    Definir puntos de interpolación dentro de las coordenadas:

    Aquí, vamos a muestrear puntos dentro de los rangos definidos por grid_x, grid_y y grid_z.

```python
# Crear una lista de puntos de interpolación dentro de la malla regular
# Vamos a tomar puntos equidistantes dentro de las coordenadas de grid_x, grid_y y grid_z

# Número de puntos en cada dirección (ajústalo según el tamaño de tus mallas)
num_points_x = len(grid_x)
num_points_y = len(grid_y)
num_points_z = len(grid_z)

# Crear los puntos de interpolación, tomados de la malla regular
interp_points = []
for i in range(num_points_x):  # Aseguramos que no excedemos el tamaño de grid_x
    for j in range(num_points_y):  # Aseguramos que no excedemos el tamaño de grid_y
        for k in range(num_points_z):  # Aseguramos que no excedemos el tamaño de grid_z
            x_prime = grid_x[i]
            y_prime = grid_y[j]
            z_prime = grid_z[k]
            interp_points.append([x_prime, y_prime, z_prime])

# Mostrar algunos puntos de interpolación
print("Puntos de interpolación:", interp_points)
```

```python
# Interpolación en los puntos seleccionados
interpolated_values = []

for point in interp_points:
    # Realizar la interpolación para cada valor en cada uno de los puntos
    interpolated_point_values = [interp(point) for interp in interpolators]
    interpolated_values.append(interpolated_point_values)

# Mostrar los resultados interpolados
for i, values in enumerate(interpolated_values):
    print(f"Punto de interpolación {i+1}: {values}")
```

```python
import numpy as np

# Suponiendo que interp_points es una lista de coordenadas 3D [(x1, y1, z1), (x2, y2, z2), ..., (x27, y27, z27)]
# Y que interpolated_values es una lista de listas con los valores interpolados

# Extraer las coordenadas 3D como entradas X
X = np.array(interp_points)  # Forma (27, 3) si hay 27 puntos de interpolación

# Extraer los valores interpolados como salidas y
y = np.array(interpolated_values)  # Forma (27, 200) si hay 200 valores por punto
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)  # Normaliza las coordenadas (x, y, z)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_normalized, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Inicializamos un modelo de RandomForest para cada uno de los 200 valores de salida
models = []
for i in range(200):
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train[:, i])  # Entrenar el modelo para el valor i
    models.append(model)

# Ahora puedes hacer predicciones por separado para cada valor
y_pred_rf = np.zeros_like(y_val)

for i in range(200):
    y_pred_rf[:, i] = models[i].predict(X_val)  # Predecir para el valor i

# Evaluación (por ejemplo, RMSE para cada valor de salida)
from sklearn.metrics import mean_squared_error

rmse_rf = np.sqrt(mean_squared_error(y_val, y_pred_rf))
print(f"RMSE (Random Forest): {rmse_rf}")
```
